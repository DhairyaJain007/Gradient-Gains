```python
team_a = df.loc[df.loc[:,"Team"] == "A"]
team_b = df.loc[df.loc[:,"Team"] == "B"]

team_a_males = team_a.loc[team_a.loc[:,"Gender"] == "M"]
team_a_females = team_a.loc[team_a.loc[:,"Gender"] == "F"]

finance = 0
hr =0
software =0



for i in df.loc[:,"Department"].to_numpy():
  if i =="Finance":
      finance += 1
  elif i =="HR":
      hr += 1
  elif i =="Software":
      software += 1

print(finance,hr,software)

print(team_a_females)
print(team_a_males)
print(team_b)
```

Implementing the Decision Tree

```python
new_team_a = []
new_team_b = []

# This does not traverse the dataframe
# for i in df:
#   print(i)

# a dataframe is a 2-d numpy array

for i in df.to_numpy():

  if i[1] == 'F':
    new_team_a.append(i)
  else:
    if i[2] == "Finance":
      new_team_b.append(i)
    else:
      if i[0] >= 41000:
        new_team_a.append(i)
      else:
        new_team_b.append(i)
```
